```python
# download data to local system
import requests
import os

data_directories = ['data/primary_label','data/full']

# make directory to store data
for data_directory in data_directories:
  try:
    os.mkdir(data_directory)
  except FileExistsError:
    print(f'Directory \'{data_directory}\' already exists.')

# download OCR file for full sheet 
url = 'https://github.com/jbest/ocr_quality/raw/main/data/datasets/BRIT/ocr/primary_label/BRIT_ocr_primary_label.zip'
r = requests.get(url, allow_redirects=True)
with open('data/BRIT_ocr_primary_label.zip', 'wb') as zip_file:
  zip_file.write(r.content)

# download OCR file for full sheet 
url = 'https://github.com/jbest/ocr_quality/raw/main/data/datasets/BRIT/ocr/full/BRIT_ocr_full.zip'
r = requests.get(url, allow_redirects=True)
with open('data/BRIT_ocr_full.zip', 'wb') as zip_file:
  zip_file.write(r.content)
```

```python
from zipfile import ZipFile

# unzip primary label image OCR
zip_path = 'data/BRIT_ocr_primary_label.zip'
data_path = 'data/primary_label/'

with ZipFile(zip_path, 'r') as zipObj:
   # Extract all the contents of zip file in current directory
   zipObj.extractall(path=data_path)

# unzip full image OCR
zip_path = 'data/BRIT_ocr_full.zip'
data_path = 'data/full/'

with ZipFile(zip_path, 'r') as zipObj:
   # Extract all the contents of zip file in current directory
   zipObj.extractall(path=data_path)
```

```python
from nostril import nonsense
from nltk.tokenize import word_tokenize

# print all primary label OCR data
for name in glob.glob('data/primary_label/*.txt'): 
    file_path = Path(name)
    #print(file_path.name, file_path.suffix, file_path.stem)
    with open(file_path) as doc_file:
      text = doc_file.read()
      words = word_tokenize(text)
      #print(text)
      for word in words:
        try:
          if nonsense(word):
            print(word)
        except:
          pass

```

```python
# find nonsense with nostril
from nostril import nonsense
from nltk.tokenize import word_tokenize

# print all primary label OCR data
for name in glob.glob('data/full/*.txt'): 
    file_path = Path(name)
    #print(file_path.name, file_path.suffix, file_path.stem)
    with open(file_path) as doc_file:
      text = doc_file.read()
      tokens = word_tokenize(text)
      #print(text)
      for token in tokens:
        try:
          if nonsense(token):
            print(token)
        except:
          pass
```

```python
# find non-dictionary words
nltk.download('words')
from nltk.corpus import words
wordlist = words.words()
```

```python
from nltk.tokenize import word_tokenize

# print all primary label OCR data
non_dict_tokens = set()
for name in glob.glob('data/full/*.txt'): 
    file_path = Path(name)
    #print(file_path.name, file_path.suffix, file_path.stem)
    with open(file_path) as doc_file:
      text = doc_file.read()
      tokens = word_tokenize(text)
      print('tokens:', len(tokens))
      #print(text)
      non_dict_count = 0
      for token in tokens:
        try:
          if token.lower() in wordlist:
            pass
            #print(token)
          else:
            #print(token)
            non_dict_count += 1
            non_dict_tokens.add(token)
        except:
          pass
      print('non_dict_count:', non_dict_count)
print(non_dict_tokens)
```

```python
# load primary label data into pandas dataframe
import pandas as pd

df = pd.DataFrame()
for name in glob.glob('data/primary_label/*.txt'): 
    file_path = Path(name)
    #print(file_path.name, file_path.suffix, file_path.stem)
    with open(file_path) as doc_file:
      text = doc_file.read()
      tokens = word_tokenize(text)
      row = {'filename': file_path.name, 'text': text, 'tokens': tokens}
      df = df.append(row, ignore_index=True)
```
